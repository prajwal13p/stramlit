```python
import pandas as pd
```

```python
df = pd.read_csv("Student-Performance-csv_T4qDx.csv")
```

```python
from sklearn.preprocessing import LabelEncoder,StandardScaler
```

```python
le = LabelEncoder()
df["Extracurricular Activities"] = le.fit_transform(df["Extracurricular Activities"])
```

```python
df.dtypes
```

```python
x = df[["Hours Studied","Previous Scores","Extracurricular Activities","Sleep Hours","Sample Question Papers Practiced"]]
y = df[["Performance Index"]]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_teat, y_train, y_test = train_test_split(x,y,train_size=0.2)
```

```python
x_train
```

```python
scalar =  StandardScaler()
x_train_scaled = scalar.fit_transform(x_train)
x_test_scaled = scalar.fit_transform(x_teat)
```

```python
from sklearn.linear_model import LinearRegression
import pickle
from sklearn.metrics import mean_squared_error,r2_score,mean_absolute_error
```

```python
model =  LinearRegression()
model.fit(x_train_scaled,y_train)
```

```python
y_pred = model.predict(x_test_scaled)
```

```python
x_test_scaled
```

```python
model.predict([[-1.15327498,  0.89115341,  1.0100505 ,  1.46427887, -1.24770427]])
```

```python
mean_absolute_error(y_test,y_pred)
```

```python
r2_score(y_test,y_pred)
```

```python
import pickle
with open("student_lr_final_model.pkl",'wb') as file:
    pickle.dump((model,scalar,le),file)
```
